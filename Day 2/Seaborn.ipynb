```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

# Distribution Plots

```python
sns.histplot(data=tips,x="total_bill",bins=20,color="skyblue",kde=True)
plt.title("Total Bill Distribution")
plt.show()
```

```python
sns.kdeplot(data=tips,x="tip",marker="x")
plt.title("Total Bill Distribution")
plt.show()
```

# Categorical Plots

```python
sns.barplot(data=tips,x="day",y='total_bill')#mean of total_bills grouped by days
plt.show()
```

```python
sns.boxplot(data=tips,x="day",y="total_bill",hue="sex",palette="deep")
plt.show()
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="day",style="smoker")
plt.show()
```

```python
sns.lineplot(data=tips,x="size",y="tip",hue="smoker")
plt.show()
```

```python
sns.pairplot(data=tips,hue="sex")
plt.show()
```

```python
corr_mat = tips.corr(numeric_only=True)
print(corr_mat)
sns.heatmap(data=corr_mat,annot=True,cmap="coolwarm",fmt=".3f")
plt.show()
```

# Titanic Dataset

```python
import pandas as pd
t_data = pd.read_csv("./train.csv")
t_data.head()
```

```python
#Task 1
sns.barplot(data=t_data,x="Pclass",y="Survived",hue="Sex")
plt.title("Survival Rate vs Class")
plt.show()
```

```python
#Task 2
sns.scatterplot(data=t_data,x="Age",y="Fare",hue="Survived",style="Sex")
plt.show()
```

```python
sns.pairplot(data=t_data,vars=["Age","Fare","Pclass"],hue="Survived",diag_kind="kde")
plt.show()
```

```python
altered_t_data = t_data.iloc[:,1:]
altered_t_data["Sex"].apply(lambda x: 0 if x=="male" else 1)
altered_t_data.head()
```

```python
corr_mat = altered_t_data.corr(numeric_only=True)
sns.heatmap(corr_mat,annot=True)
plt.show()
```
